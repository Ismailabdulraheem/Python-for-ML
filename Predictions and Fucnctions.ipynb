### Create train variables

```python
x_train = np.array([3.0,1.0,2.0])
y_train = np.array([300.0,500.0,700.0])
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

### Adjust the Model Parameters

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

### Compute the model output

```python
def main_model(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
f_wb = main_model(x_train, w, b)
plt.plot(x_train, f_wb, c = 'b')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')
```

### Compute Cost

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0.0
    for i in range(m):
        f_wb = w * x[i] + b
        cost += (f_wb - y[i]) ** 2
    total_cost = 1.0 / (2 * m) * cost
    return total_cost
```

 Compute Gradient

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x[i] + b 
        dj_dw_i = (f_wb - y[i]) * x[i] 
        dj_db_i = f_wb - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```
