Logistic Cost Function

The cost function is

$$ J(\mathbf{w},b) = \frac{1}{m} \sum_{i=0}^{m-1} \left[ loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) \right]$$

where
* $loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)})$ is

    $$loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = -y^{(i)} \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)$$
    
and
$$f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(z^{(i)})$$
$$z^{(i)} = \mathbf{w} \cdot \mathbf{x}^{(i)}+ b$$
$$g(z^{(i)}) = \frac{1}{1+e^{-z^{(i)}}}$$

```python
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        z = np.dot(X[i], w) + b
        f_wb = sigmoid(z)
        cost = cost - y[i] * np.log(f_wb) - (1 - y[i]) * np.log(1 - f_wb)
    cost = cost / m
    return cost
w = np.array([1.0, 1.0])
b = -3.0
print(compute_cost_logistic(X_train, y_train, w, b))
```

From the charts, we can see that the decision boundary with b = -3, $w_0$ = 1, $w_1$ = 1 fits the data better than the other one. This indicates that the first set of parameters provides a better model for the given dataset.
