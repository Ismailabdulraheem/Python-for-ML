Import all modules needed

```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import plot_data
```

Build dataset

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
from tkinter import font
from numpy import positive


positive = y_train == 1
negative = y_train == 0

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
#plot 1D dataset
ax[0].scatter(x_train[negative], y_train[negative], marker='o', color='red', label='Negative class (0)', s = 100)
ax[0].scatter(x_train[positive], y_train[positive], marker='x', color='blue', label='Positive class (1)', s = 100)
ax[0].set_title('1D Classification Dataset')
ax[0].set_xlabel('x', fontsize=14)
ax[0].set_ylabel('y', fontsize=14)
ax[0].set_ylim(-0.08, 1.1)
ax[0].legend(fontsize=12)

#plot 2D dataset (2 variables)
plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_title('2D Classification Dataset')
ax[1].set_xlabel('x1', fontsize=14)
ax[1].set_ylabel('$x_2$', fontsize=14)
ax[1].legend(fontsize=12)
plt.tight_layout()
plt.show()
```
