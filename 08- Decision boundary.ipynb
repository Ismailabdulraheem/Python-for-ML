````python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import plot_data, sigmoid, draw_vthresh
````

Create a dataset

````python
X = np.array([[0.50, 1.50], [1.00, 1.00], [1.50, 0.50], [3.00, 0.50], [2.00, 2.00], [1.00, 2.50]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1, 1)
````

````python
fig, ax = plt.subplots(1, 1, figsize=(5,5))
plot_data(X, y, ax)
ax.axis([0.0, 4.0, 0.0, 4.0])
ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')
plt.show()
````

````python
z = np.arange(-10,11)

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(z, sigmoid(z), color='blue')
ax.set_title('Sigmoid function')
ax.set_xlabel('z')
ax.set_ylabel('$\sigma(z)$')
draw_vthresh(ax, 0)
````

Plotting decision boundary

The model has the form
 $f(\mathbf{x}) = g(-3 + x_0+x_1)$

 Then $y=1$ if $-3 + x_0+x_1 >= 0$
      $$-3 + x_0+x_1 = 0$$
      $$x_1 = 3 - x_0$$

````python
x0 = np.arange(0,6)
x1 = 3 - x0
fig, ax = plt.subplots(1, 1, figsize=(5,5))
plot_data(X, y, ax)
ax.plot(x0, x1, color='green') #for decision boundary
ax.fill_between(x0, x1, alpha = 0.2)
ax.axis([0.0, 4.0, 0.0, 4.0])
ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')
plt.show()
````
